MacMahon's Partitions Function


```python
def MacMahon(a, n):
    if a <= 0 or n <= 0:
        print("Error: 'a' and 'n' must be positive integers.")
        return 0
    total_sum = [0]
    _find_partitions_recursive(n, a, 1, [], total_sum)
    return total_sum[0]

def _find_partitions_recursive(target_n, blocks_left, min_s, heights, total_sum):
    if blocks_left == 1:
        for s in range(min_s, target_n + 1):
            if target_n % s == 0:
                d = target_n // s
                product = d
                for h in heights:
                    product *= h
                total_sum[0] += product
        return

    for s in range(min_s, target_n):
        for d in range(1, (target_n // s) + 1):
            remaining_n = target_n - (d * s)
            min_possible_sum = sum(range(s + 1, s + 1 + (blocks_left - 1)))

            if remaining_n >= min_possible_sum:
                _find_partitions_recursive(
                    remaining_n,
                    blocks_left - 1,
                    s + 1,
                    heights + [d],
                    total_sum
                )
```

Check M_1(n)

```python
for n in range(1, 100):
  print(n, (MacMahon(1, n)))
```

Check M_2(n)

```python
for n in range(1, 100):
  print(n, (MacMahon(2, n)))
```

Some polynomials detect primes

```python
f1 = lambda n: (n**2 - 3*n + 2)*(MacMahon(1, n))
f2 = lambda n: 8*MacMahon(2, n)
f = lambda n: f1(n) - f2(n)
```

```python
for n in range(1, 101):
    print(n, f(n))
```

```python
g1 = lambda n: (3*n**3 - 13*n**2 + 18*n - 8)*(MacMahon(1, n))
g2 = lambda n: (12*n**2 - 120*n + 212)*(MacMahon(2,n))
g3 = lambda n: 960*(MacMahon(3,n))
g = lambda n: g1(n) + g2(n) - g3(n)
```

```python
for n in range(1, 101):
    print(n, g(n))
```

```python
h1 = lambda n: (25*n**4 - 171*n**3 + 423*n**2 - 447*n + 170)*(MacMahon(1, n))
h2 = lambda n: (300*n**3 - 3554*n**2 + 12900*n - 14990)*(MacMahon(2,n))
h3 = lambda n: (2400*n**2 - 60480*n +214080)*(MacMahon(3,n))
h4 = lambda n: 725760*(MacMahon(4,n))
h = lambda n: h1(n) + h2(n) + h3(n) - h4(n)
```

```python
for n in range(1, 101):
    print(n, h(n))
```

Some Congruences

M_2(5n + 2) is divisible by 5

```python
for n in range(1, 101):
  if (MacMahon(2, 5*n+2) % 5 == 0):
    print(n, (MacMahon(2, 5*n+2)), "Divisible by 5")
```

M_3(7n + 3) and M_3(7n + 5) is divisible by 7

```python
for n in range(1, 51):
  if (MacMahon(3, 7*n+3) % 7 == 0):
    print(n, (MacMahon(3, 7*n+3)), "Divisible by 7")
```

```python
for n in range(1, 51):
  if (MacMahon(3, 7*n+5) % 7 == 0):
    print(n, (MacMahon(3, 7*n+5)), "Divisible by 7")
```
